```python
import pandas as pd
```

```python
covid_df = pd.read_csv("singapore_covid19_data.csv")
```

```python
covid_df['Date'] = pd.to_datetime(covid_df["Date"]).dt.date

covid_df = covid_df.sort_values('Date').set_index('Date')
```

```python
covid_df.head()
```

### Confirmed Cases

```python
covid_df.Confirmed.plot.line(title='Confirmed Cases', figsize=(15,8), grid=True, color='blue', legend=True)
```

### Active Cases

```python
covid_df.Active.plot.line(title='Active Cases', figsize=(15,8), grid=True, color='red', legend=True)
```

### Recovered

```python
covid_df.Recovered.plot.line(title='Recovered Cases', figsize=(15,8), grid=True, color='green', legend=True)
```

### Deaths Cases

```python
covid_df.Deaths.plot.line(title='Deaths Cases', figsize=(15,8), grid=True, color='black', legend=True)
```
